Part one
--------

I actually spent all day on this before realizing it could be solved without a computer at all.

Stating the obvious:

* Because our initial position is $\left(0, 0\right)$ and the target's minimum coordinate along the $x$-axis is greater than zero, our initial velocity along the $x$-axis must be positive.

* That means that our velocity along the $x$-axis across the probe's entire trajectory is either positive or zero.

* And that means that the probe's _position_ vector can only have a finite number of values for its $x$-coordinate.

Also obvious, but it took me hours of math before fully grokking its implications:

**Observation 1**: The probe's motion along the $y$-axis is independent from its motion along the $x$-axis. We're looking for the maximum height, so we only need to consider velocity and position along the $y$-axis — since the problem has a solution, we know that there's some value for the initial velocity's $x$-component that lands the probe in the target area, _but we don't need to find it_.

From now on, talking about velocity and position, let's assume we're working in a single dimension along the $y$-axis.

The target ranges over the closed interval $\left[T_{min}, T_{max}\right]$, where $T_{max} < 0$.

---

**Observation 2** is about the important points in the trajectory of the probe. Given initial velocity $v_0$, then the probe's position is zero at times $t = 0$ and $t = 2v_0 + 1$.

The probe's velocity changes by -1 each unit of time. So $v\left(t\right) = v_0 - t$. 

It takes $v_0$ units of time for the probe's velocity to change from $v_0$ to 0. The probe's velocity is zero for one unit of time, and then another $v_0$ units for its velocity to change from zero to $-v_0$. Total time is $2v_0 + 1$.

We can add all of those velocity vectors to see that each vector pointing in the positive direction has a corresponding vector in the negative direction. They all cancel out so the net velocity is zero, and the net displacement is zero.

---

**Observation 3**: At time $2v_0 + 1$, the probe's velocity is $-v_0 - 1$. That means its _position_ at time $2v_0 + 2 = -v_0 - 1$ (because its position at $t = 2v_0 + 1$ is zero).

This puts an upper bound on our initial velocity, which is what we want in order to find the upper bound on the probe's height. Notice that if $v_0 > -T_{min} + 1$, then we have the following:

* At time $2v_0 + 1$, we have $y = 0 > T_{max}$ (from above).

* At time $2v_0 + 2$, we have $y = -v_0 - 1$. But if $v_0 > -T_{min} + 1$ then we have: $$v_0 - 1 > -T_{min}$$ and $$-v_0 + 1 < T_{min}$$ (multiplying by -1 switches the signs) and finally $$-v_0 - 1 < -v_0 + 1 < T_{min}$$ is outside the target bounds.

The upshot of this whole thing is that $v_{max} = -T_{min} + 1$.

---

**Observation 4**: The maximum height of the probe with initial velocity $v_0$ (where $v_0 > 0$) is the $v_0$-th triangular number.

In Observation 2, we noticed that at time $t = v_0$, the velocity of the probe had take all (and only) values that are natural numbers less than or equal to $v_0$. Each velocity lasted for a duration of one time unit, so the position vector is simply the sum of those velocity vectors.

---

So without writing a line of Python code (even though I _did_ write a lot of Python code), given the example problem:

```
target area: x=20..30, y=-10..-5
```

We see that $$\begin{aligned} y_{min} &= -10 \\ ⇒ v_{y_{max}} &= -(-10 + 1) = 9 \\ ⇒ y_{max} &= \frac{9 \cdot 10}{2} = 45 \end{aligned}$$.
